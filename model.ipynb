```python
# Check for missing values in decreasing order
df.isnull().mean().sort_values(ascending=False)
```

```python
threshold = 0.3
missing_ratio = df.isnull().mean()
df = df.loc[:, missing_ratio < threshold]
```

```python
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
missing_ratio_num = df[num_cols].isnull().mean()
```

```python
knn_threshold = 0.1
knn_cols = missing_ratio_num[missing_ratio_num > knn_threshold].index.tolist()
simple_cols = missing_ratio_num[missing_ratio_num <= knn_threshold].index.tolist()
skewness = df[simple_cols].skew()
median_cols = skewness[skewness.abs() > 0.5].index.tolist()
mean_cols = skewness[skewness.abs() <= 0.5].index.tolist()
```

```python
median_imputer = SimpleImputer(strategy='median')
mean_imputer = SimpleImputer(strategy='mean')
knn_imputer = KNNImputer(n_neighbors=5)
mode_imputer = SimpleImputer(strategy='most_frequent')
```

```python
transformers = []

if median_cols:
    transformers.append(('median_imputer', median_imputer, median_cols))
if mean_cols:
    transformers.append(('mean_imputer', mean_imputer, mean_cols))
if knn_cols:
    transformers.append(('knn_imputer', knn_imputer, knn_cols))
if len(cat_cols) > 0:
    transformers.append(('mode_imputer', mode_imputer, cat_cols))

imputer = ColumnTransformer(transformers=transformers, remainder='passthrough')
```

```python
# Fit transform
df_imputed = imputer.fit_transform(df)

# Convert back to DataFrame (adjust columns order as needed)
imputed_cols = median_cols + mean_cols + knn_cols + list(cat_cols) + \
               [col for col in df.columns if col not in median_cols + mean_cols + knn_cols + list(cat_cols)]

df_imputed = pd.DataFrame(df_imputed, columns=imputed_cols)
```

```python
df = df_imputed
```
